<h2 style="text-align: left; font-size: 32px; background-color: #f59d5a; padding: 10px; border-radius: 10px;"> What is COVID-19 ? </h2>
<p style="text-align: left; font-size: 25px; padding: 5px; border-radius: 10px;">
    COVID-19, short for "Coronavirus Disease 2019," is an infectious disease caused by the severe acute respiratory syndrome coronavirus 2 (SARS-CoV-2). It was first identified in December 2019 in the city of Wuhan, Hubei province, China, and has since developed into a global pandemic.
</p>


<h2 style="text-align: left; font-size: 32px; background-color: #f59d5a; padding: 10px; border-radius: 10px;">  Importing necessary libraries.... </h2>


```python
import pandas as pd
import plotly.offline as pyo 
import plotly.express as px
import plotly.io as pio 
import plotly.graph_objects as go
import matplotlib.pyplot as plt
```

<h2 style="text-align: left; font-size: 32px; background-color: #f59d5a; padding: 10px; border-radius: 10px;">  Importing Data .... </h2>

```python
df = pd.read_csv('/kaggle/input/real-time-latest-covid-19-dataset/covid.csv')
df.head()
```

<h2 style="text-align: left; font-size: 32px; background-color: #f59d5a; padding: 10px; border-radius: 10px;">  Data Analysis/EDA .... </h2>


```python
fig_total_cases = px.choropleth(df, 
                                locations='Country/other', 
                                locationmode='country names', 
                                color='Total_cases', 
                                color_continuous_scale='Oranges',
                                title='World Map: Total COVID-19 Cases')
fig_total_cases.show()
```

```python
fig_total_deaths = px.choropleth(df, 
                                 locations='Country/other', 
                                 locationmode='country names', 
                                 color='Total_deaths', 
                                 color_continuous_scale='Reds',
                                 title='World Map: Total COVID-19 Deaths')
fig_total_deaths.show()
```

```python
fig_active_cases = px.choropleth(df, 
                                 locations='Country/other', 
                                 locationmode='country names', 
                                 color='Active_cases', 
                                 color_continuous_scale='Blues',
                                 title='World Map: Active COVID-19 Cases')
fig_active_cases.show()
```

```python
fig_total_recoveries = px.choropleth(df, 
                                     locations='Country/other', 
                                     locationmode='country names', 
                                     color='Total_recovered', 
                                     color_continuous_scale='Greens',
                                     title='World Map: Total COVID-19 Recoveries')
fig_total_recoveries.show()
```

```python
fig_serious_cases = px.choropleth(df, 
                                  locations='Country/other', 
                                  locationmode='country names', 
                                  color='Serious_cases', 
                                  color_continuous_scale='Plasma',  
                                  title='World Map: Serious COVID-19 Cases')
fig_serious_cases.show()
```

```python
fig_total_tests = px.choropleth(df, 
                                locations='Country/other', 
                                locationmode='country names', 
                                color='Total_tests', 
                                color_continuous_scale='Viridis',  
                                title='World Map: Total COVID-19 Tests')
fig_total_tests.show()
```

```python

top_10_cases = df.sort_values(by='Total_cases', ascending=False).head(10)

fig = px.bar(top_10_cases, x='Country/other', y='Total_cases', title='Top 10 Countries with the Most COVID-19 Cases')

bar_color = '#1f77b4'
fig.update_traces(marker=dict(color=bar_color))

fig.update_layout(plot_bgcolor='#5e6a6f')

fig.show()

```

```python
top_10_deaths = df.sort_values(by='Total_deaths', ascending=False).head(10)
fig = px.bar(top_10_deaths, x='Country/other', y='Total_deaths', title='Top 10 Countries with the Most COVID-19 Deaths')
fig.show()
```

```python
top_10_recoveries = df.sort_values(by='Total_recovered', ascending=False).head(10)

fig = px.bar(
    top_10_recoveries,
    x='Country/other',
    y='Total_recovered',
    title='Top 10 Countries with the Most COVID-19 Recoveries',
)

fig.update_layout(
    plot_bgcolor='Black',  
    paper_bgcolor='lightgray', 
    yaxis=dict(showgrid=True, gridcolor='lightgray'),  
)

fig.show()

```

```python
top_10_recoveries = df.sort_values(by='Total_recovered', ascending=False).head(10)

fig = px.bar(top_10_recoveries, x='Country/other', y='Total_recovered', title='Top 10 Countries with the Most COVID-19 Recoveries')

fig.update_layout(
    title_font_color="white",  
    plot_bgcolor="black",      
    paper_bgcolor="lightgray"       
)

fig.update_traces(marker_color='limegreen',textfont_color='white')  

fig.show()
```

```python
df['Death_rate'] = (df['Total_deaths'] / df['Total_cases']) * 100
df['Recovery_rate'] = (df['Total_recovered'] / df['Total_cases']) * 100
```

```python

top_10_death_rate = df.sort_values(by='Death_rate', ascending=False).head(10)

fig = px.line(top_10_death_rate, x='Country/other', y='Death_rate', title='Top 10 Countries with the Highest COVID-19 Death Rate')

fig.update_layout(
    plot_bgcolor='#cbd3d7',  
    xaxis_showgrid=False,     
    yaxis_showgrid=False     
)


fig.show()
```

```python


top_10_recovery_rate = df.sort_values(by='Recovery_rate', ascending=False).head(10)

fig = px.bar(top_10_recovery_rate, x='Country/other', y='Recovery_rate', title='Top 10 Countries with the Highest COVID-19 Recovery Rate')

fig.update_layout(
    title_font_color="white",  
    plot_bgcolor="black",      
    paper_bgcolor="lightgray"       
)

fig.update_traces(marker_color='dodgerblue')  

fig.show()
```

```python
top_10_active_cases = df.sort_values(by='Active_cases', ascending=False).head(10)
fig = px.bar(top_10_active_cases, x='Country/other', y='Active_cases', title='Top 10 Countries with the Most Active COVID-19 Cases')
fig.show()
```

```python


top_10_serious_cases = df.sort_values(by='Serious_cases', ascending=False).head(10)

labels = top_10_serious_cases['Country/other']
values = top_10_serious_cases['Serious_cases']

fig = go.Figure(data=[go.Pie(labels=labels, values=values)])

fig.update_traces(textinfo='percent+label', 
                  hoverinfo='label+percent', 
                  textfont_size=15,
                  )

fig.update_layout(title=' Current Top 10 Countries with the Most Serious COVID-19 Cases',
                  title_x=0.5, 
                  margin=dict(l=0, r=0, t=50, b=0)) 

fig.show()
```

```python
top_10_tests = df.sort_values(by='Total_tests', ascending=False).head(10)

fig = px.bar(top_10_tests, x='Country/other', y='Total_tests', title='Top 10 Countries with the Most COVID-19 Tests Conducted')

fig.update_layout(
    title_font_color="white",  
    plot_bgcolor="black",      
    paper_bgcolor="lightgray"      
)

fig.update_traces(marker_color='royalblue') 

fig.show()
```

##### Correlation between COVID-19 cases and population, for the top 10 countries

```python
top_10_population = df.sort_values(by='Total_cases', ascending=False).head(10)
correlation_cases_population = top_10_population[['Total_cases', 'Population']].corr()
correlation_cases_population
```

 ##### Correlation between COVID-19 deaths and population, for the top 10 countries

```python
correlation_deaths_population = top_10_population[['Total_deaths', 'Population']].corr()
correlation_deaths_population
```

####  Correlation between COVID-19 recoveries and population, for the top 10 countries

```python
correlation_recoveries_population = top_10_population[['Total_recovered', 'Population']].corr()
correlation_recoveries_population
```

```python
top_10_recoveries = df.sort_values(by='Total_recovered', ascending=False).head(5)
fig = px.scatter(top_10_recoveries, x='Total_recovered', y='Population', size='Total_recovered', text='Country/other', title='Top 10 Countries: COVID-19 Recoveries vs Population')
fig.update_traces(marker=dict(sizemode='diameter', opacity=0.4))
fig.show()
```
